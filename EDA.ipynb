```python
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV data into a NumPy array
data = np.loadtxt('Ba133a.csv', delimiter=',', skiprows=1)

# Separate the data into channels and counts
channels = data[:,0]
counts = data[:,1]

# Plot the spectrum
plt.plot(channels, counts)
plt.xlabel('Channel')
plt.ylabel('Counts')
plt.title('Ba-133 Spectrum')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from each file into a list
files = ['Ba133a.csv', 'Ba133b.csv', 'Ba133c.csv', 'Ba133d.csv']
data = [np.loadtxt(f, delimiter=',', skiprows=1) for f in files]

# Plot each spectrum in a separate subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.flatten()
for i in range(len(files)):
    channels = data[i][:,0]
    counts = data[i][:,1]
    axs[i].plot(channels, counts)
    axs[i].set_xlabel('Channel')
    axs[i].set_ylabel('Counts')
    axs[i].set_title('Ba-133 Spectrum (' + files[i] + ')')
plt.tight_layout()
plt.show()
```

Dari visualisasi di atas, dapat dilihat bahwa:
- Spektrum dari elemen yang sama masing-masing memiliki bentuk yang sama 
- Counts/ cacahan yang diperoleh menjadi masalah di sini karena nilainya 2x lebih besar daripada pengukuran sebelumnya
- Untuk data dengan dua elemen tidak terlihat adanya pattern bahwa spektrum tersebut merupakan kombinasi dari kedua elemen tersebut
